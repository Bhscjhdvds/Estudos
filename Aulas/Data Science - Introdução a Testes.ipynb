```python
import pandas as pd
```

```python
tmdb = pd.read_csv('tmdb_5000_movies.csv')
tmdb.head()
```

```python
tmdb.describe()
```

```python
import seaborn as sns
```

```python
ax = sns.histplot(x="vote_average", data=tmdb, kde=True)
ax.set(xlabel="Nota média", ylabel="Frenquencia")
ax.set_title("Média de votos em filmes no TMDB 5000")
```

```python
ax = sns.histplot(x="vote_average", data=tmdb, stat="probability" , kde=True)
ax.set(xlabel="Nota média", ylabel="Densidade")
ax.set_title("Média de votos em filmes no TMDB 5000")
```

```python
ax = sns.boxplot(data=tmdb, x="vote_average")
ax.set(xlabel="Nota média do filme ")
ax.set_title("Distribuição de notas médias dos filmes do TMDB 5000")
```

```python
tmdb.query("vote_average == 0")
```

```python
tmdb_mais_de_10_votos = tmdb.query("vote_count >= 10")
tmdb_mais_de_10_votos.describe()
```

```python
ax = sns.histplot(x="vote_average", data=tmdb_mais_de_10_votos, kde=True)
ax.set(xlabel="Nota média", ylabel="Frenquencia")
ax.set_title("Média de votos em filmes no TMDB 5000")
```

```python
ax = sns.histplot(x="vote_average", data=tmdb_mais_de_10_votos, stat="probability" , kde=True)
ax.set(xlabel="Nota média", ylabel="Densidade")
ax.set_title("Média de votos em filmes no TMDB 5000")
```

```python
notas = pd.read_csv('ratings.csv')
notas.head()
```

```python
nota_media_por_filme = notas.groupby('movieId').mean()["rating"]
nota_media_por_filme.head()
```

```python
ax = sns.histplot(x=nota_media_por_filme.values, kde=True)
ax.set(xlabel="Nota média", ylabel="Frenquencia")
ax.set_title("Média de votos em filmes no TMDB 5000")
```

```python
ax = sns.histplot(x=nota_media_por_filme.values, stat="probability" , kde=True)
ax.set(xlabel="Nota média", ylabel="Densidade")
ax.set_title("Média de votos em filmes no TMDB 5000")
```
